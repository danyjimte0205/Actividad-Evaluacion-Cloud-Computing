```python
import pandas as pd
import json
import requests

# Cargar los datos de prueba
data = pd.read_csv("salesLT_costumer.csv").drop(["ModifiedDate"], axis = 1)
data_dict = data.to_dict(orient = 'list')

# Convertir datos al formato JSON esperado por la API
data_json = json.dumps({"data": [data_dict]})
```

```python
# Leer la URI de la API desde un archivo JSON
with open("uri.json", "r") as suri:
    scoring_uri = json.load(suri)["URI"][0]

# Definir encabezados para la solicitud HTTP
headers = {"Content-Type": "application/json"}

# Hacer la petición a la API de Azure
response = requests.post(scoring_uri, data=data_json, headers=headers)
```

```python
# Manejo de la respuesta
if response.status_code == 200:
    result = json.loads(response.json())
    print(result)
    data["ModifiedDate"] = result
    display(data)
else:
    print(f"Error: {response.text}")
```
